```python
# IMPORTS
import os
import time
import requests
import hmac
from hashlib import sha256
```

```python
# CONSTANTS
from dotenv import load_dotenv

load_dotenv()

APIURL = os.getenv("API_URL")
APIKEY = os.getenv("API_KEY")
SECRETKEY = os.getenv("SECRET_KEY")
```

```python
# USDT-M Perp Futures symbols


def demo():
    payload = {}
    path = "/openApi/swap/v2/quote/contracts"
    method = "GET"
    paramsMap = {}
    paramsStr = parseParam(paramsMap)
    return send_request(method, path, paramsStr, payload)


def get_sign(api_secret, payload):
    signature = hmac.new(
        api_secret.encode("utf-8"), payload.encode("utf-8"), digestmod=sha256
    ).hexdigest()
    print("sign=" + signature)
    return signature


def send_request(method, path, urlpa, payload):
    url = "%s%s?%s&signature=%s" % (APIURL, path, urlpa, get_sign(SECRETKEY, urlpa))
    print(url)
    headers = {
        "X-BX-APIKEY": APIKEY,
    }
    response = requests.request(method, url, headers=headers, data=payload)
    return response.text


def parseParam(paramsMap):
    sortedKeys = sorted(paramsMap)
    paramsStr = "&".join(["%s=%s" % (x, paramsMap[x]) for x in sortedKeys])
    if paramsStr != "":
        return paramsStr + "&timestamp=" + str(int(time.time() * 1000))
    else:
        return paramsStr + "timestamp=" + str(int(time.time() * 1000))


if __name__ == "__main__":
    print("demo:", demo())
```

```python
# Kline/Candlestick Data Slope
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Datos proporcionados
data = [
    # {'close': '62885.0', 'time': 1728326880000},
    # {'close': '62891.9', 'time': 1728326820000},
    # {'close': '63030.5', 'time': 1728326760000},
    # {'close': '63157.5', 'time': 1728326700000},
    # {'close': '63173.9', 'time': 1728326640000},
    # {'close': '63212.4', 'time': 1728326580000},
    # {'close': '63200.6', 'time': 1728326520000}
]


def demo():
    payload = {}
    path = "/openApi/swap/v3/quote/klines"
    method = "GET"
    paramsMap = {
        "symbol": "BTC-USDT",
        "interval": "1m",
        "limit": "200",
        # "startTime": str(int(time.time() * 1000))
    }
    paramsStr = parseParam(paramsMap)
    return send_request(method, path, paramsStr, payload)


def get_sign(api_secret, payload):
    signature = hmac.new(
        api_secret.encode("utf-8"), payload.encode("utf-8"), digestmod=sha256
    ).hexdigest()
    print("sign=" + signature)
    return signature


def send_request(method, path, urlpa, payload):
    url = "%s%s?%s&signature=%s" % (APIURL, path, urlpa, get_sign(SECRETKEY, urlpa))
    print(url)
    headers = {
        "X-BX-APIKEY": APIKEY,
    }
    response = requests.request(method, url, headers=headers, data=payload)
    # return response.text
    data = response.json()["data"]
    if response.json()["code"] != 0:
        print(response.text)
    return data


def parseParam(paramsMap):
    sortedKeys = sorted(paramsMap)
    paramsStr = "&".join(["%s=%s" % (x, paramsMap[x]) for x in sortedKeys])
    if paramsStr != "":
        return paramsStr + "&timestamp=" + str(int(time.time() * 1000))
    else:
        return paramsStr + "timestamp=" + str(int(time.time() * 1000))


if __name__ == "__main__":
    data = demo()
    print("demo:", data)

    # Obtener los últimos 7 valores
    limit = 20
    last = limit * -1
    data = data[last:]
    # Convertir a DataFrame
    df = pd.DataFrame(data)
    print(df)
    # Convertir las columnas 'close' a valores numéricos y 'time' a un formato manejable
    df["close"] = pd.to_numeric(df["close"])
    df["time"] = pd.to_datetime(df["time"], unit="ms")

    # Extraer el valor de 'time' como número ordinal para calcular la pendiente
    # df['time_ordinal'] = pd.to_datetime(df['time']).map(pd.Timestamp.toordinal)

    # Convertir la columna 'time' a segundos desde el primer timestamp
    df["time_ms"] = (df["time"] - df["time"].min()).dt.total_seconds()

    # Regresión lineal para calcular la pendiente usando tiempo en segundos
    # slope, intercept, r_value, p_value, std_err = stats.linregress(df['time_ordinal'], df['close'])
    slope, intercept, r_value, p_value, std_err = stats.linregress(
        df["time_ms"], df["close"]
    )

    # Graficar los datos de precios
    plt.figure(figsize=(10, 6))
    plt.plot(df["time"], df["close"], label="Precio BTC", marker="o")

    # Graficar la línea de tendencia (recta de la regresión lineal)
    # plt.plot(df['time'], intercept + slope * df['time_ordinal'], 'r', label=f'Línea de tendencia (Pendiente: {slope:.2f})')
    plt.plot(
        df["time"],
        intercept + slope * df["time_ms"],
        "r",
        label=f"Línea de tendencia (Pendiente: {slope:.2f})",
    )

    # Configuración del gráfico
    plt.xlabel("Fecha")
    plt.ylabel("Precio BTC (close)")
    plt.title("Precio de BTC y su tendencia")
    plt.legend()
    plt.grid(True)

    # Mostrar la gráfica
    plt.show()

    # Resultados de la regresión lineal
    print(f"Pendiente: {slope}")
    print(f"Intercepto: {intercept}")
    print(f"R-cuadrado: {r_value**2}")
    print(f"P-valor: {p_value}")
    print(f"Error estándar: {std_err}")

    # Extraer los precios de cierre (close) y los timestamps
    close_prices = np.array([float(item["close"]) for item in data])
    timestamps = np.array([item["time"] for item in data])

    # print("Precios de cierre:", close_prices)
    # print("Timestamps:", timestamps)
    # Convertir los timestamps a segundos desde el primer valor
    time_seconds = (
        timestamps - timestamps.min()
    ) / 1000  # Convertir milisegundos a segundos

    # Cálculo manual de la pendiente usando NumPy
    # Aplicamos la fórmula de la regresión lineal: y = mx + b
    A = np.vstack([time_seconds, np.ones(len(time_seconds))]).T
    slope, intercept = np.linalg.lstsq(A, close_prices, rcond=None)[0]

    # Mostrar la pendiente
    print(f"Pendiente: {slope}")
```
